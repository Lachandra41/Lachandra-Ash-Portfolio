```python
#Data Visualization
```

```python
#Display the number of rating.
plt.hist(df.rating)
plt.title('Number of Coffee Ratings') 
plt.xlabel('Coffee Ratings')
plt.ylabel('Coffee Ratings Count')
plt.show()
```

```python
#Plot rating and number of coffee products.
plt.hist(df.rating, bins = np.arange(3,5,0.25))
plt.title('Ratings and Number of Coffee Products') 
plt.xlabel('Coffee Ratings')
plt.ylabel('Count of Coffee Products')
plt.show()
```

```python
#Display the coffee prices and weight formated to grams.
sns.lineplot(x ='weight_formatted_to_grams', y = 'price', data= df);
plt.title('Coffee Price and Weight Formatted to Grams') 
plt.show()
```

```python
#Display the price and rating of the coffees.
sns.barplot(x='rating', y = 'price', data= df);
```

```python
#Display the coffee ratings versus reviews.
sns.barplot(x='rating', y = 'reviews', data= df);
```
